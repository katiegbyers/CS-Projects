## Transformations

A transformer is an object that cleans, reduces, expands, or generates features. A transformer's fit method, will learn parameters.

The transform method takes the learned parameters and transforms any inputted new data. This means taking a vector of categorical values and transforming the data into a matrix where each row has at most one non-zero value (they may all be zero if this vector contains a category that was previously unseen).

fit_transform simply learns from and transforms the input data all in one go

Since I might want to perform different transformations on different columns, I've provided a ColumnSelector class.

```python
from sklearn.base import BaseEstimator, TransformerMixin

class ColumnSelector(BaseEstimator, TransformerMixin):
    """
    Transformer that extracts a column of a data frame
    
    Example Usage
    >> data = pd.DataFrame({'a': [1, 2, 3, 4],
                            'b': [5, 6, 7, 8],
                            'c': [9, 10, 11, 12]})
    >> cs = ColumnSelector(cols=['a', 'b'])
    >> data['a'] == cs.transform(data)

    Parameters
    ----------
    col : list of strings, required
        The name(s) corresponding to the desired column of a DataFrame.
    """
    
    def __init__(self, cols):
        self.cols = cols

    def fit(self, X, y=None):
        """
        Returns itself, nothing to be fit
        """
        return self

    def transform(self, X, y=None):
        """
        Returns the desired column as a matrix
        """
        return X.as_matrix(self.cols)
```

## Pipeline

At this point, we will formalize our data cleaning, extraction, transformation, and training all into an abstraction called a Pipeline. In a nutshell, a pipeline is the recipe for going from a clean but untransformed data set to a trained model. Below I extracted polynomial features from each woman's form of standard living and then fit a linear model.

```python
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures

pipeline1 = Pipeline([
    ('selector', ColumnSelector(['standard_living'])), 
    ('poly_feats', PolynomialFeatures(3, include_bias=False)),
    ('lm', lm.LinearRegression(fit_intercept=False))
])

pipeline1.fit(data, data[['contraceptive']])

print("Training RMSE:",
      (np.mean((pipeline1.predict(data) - data[['contraceptive']])**2)**(.5)).item())
```

In addition to our training error, I want to be able to calculate validation error using cross validation. Below, I calculated the validation error for our initial pipeline, using 10-fold cross-validation.

```python
from sklearn.metrics import make_scorer
from sklearn.model_selection import cross_val_score

def score_func(y, y_pred, **kwargs): 
    return np.mean((y - y_pred)**2)**0.5
scorer = make_scorer(score_func)

cv_scores = cross_val_score(pipeline1, data, data[['contraceptive']], cv=10, scoring=scorer)
print("Validation RMSE:", np.mean(cv_scores))
```

Of course we wouldn't want to just use one predictor. To append more features, we can use FeatureUnion, which combines several transformers into a MEGATRANSFORMER, which outputs a concatenation of the output of its constituents. Below I combined the polynomial lift of standard_living and wife_age.

```python
pipeline2 = Pipeline([
    ('union', FeatureUnion(n_jobs=1,  transformer_list=[
        ('poly_sd_living', Pipeline([
            ('selector', ColumnSelector(['standard_living'])), 
            ('poly_feats', PolynomialFeatures(3, include_bias=False))
        ])),
        
        ('poly_media_exposure', Pipeline([
            ('selector', ColumnSelector(['media_exposure'])), 
            ('poly_feats', PolynomialFeatures(3, include_bias=False)),
        ]))
    ])),
        
    ('lm', lm.LinearRegression(fit_intercept=False))
])

pipeline2.fit(data, data[['contraceptive']])

print("Training RMSE:",
      (np.mean((pipeline2.predict(data) - data[['contraceptive']])**2)**(.5)).item())

cv_scores = cross_val_score(pipeline2, data, data[['contraceptive']], cv=10, scoring=scorer)
print("Validation RMSE:", np.mean(cv_scores))
```

I measured the quality of this model by calculating the Root-Mean-Square Error (RMSE) between the predicted contraceptive methods and the true contraceptive methods. 

```python
def rmse(actual, predicted):
    """
    Calculates RMSE from actual and predicted values
    Input:
      actual (1D array): vector of actual values
      predicted (1D array): vector of predicted/fitted values
    Output:
      a float, the root-mean square error
    """
    return math.sqrt((np.sum((actual - predicted)**2)) / len(actual))
```

```python
training_error = rmse(y_train, y_fitted)
test_error = rmse(y_test, y_predicted)
(training_error, test_error)
```
